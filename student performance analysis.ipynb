```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "Name": ["Aman", "Riya", "Rahul", "Sneha", "Arjun", "Priya", "Karan", "Neha"],
    "Math": [78, 85, 90, 65, 88, 92, 70, 60],
    "Science": [80, 88, 85, 70, 90, 95, 75, 65],
    "English": [75, 82, 88, 68, 85, 90, 72, 66]
}

import pandas as pd
df = pd.DataFrame(data)
df
```

```python
# Check missing values
print(df.isnull().sum())

# Remove duplicates
df = df.drop_duplicates()

# Check data types
print(df.dtypes)
```

```python
# Average marks
print("Average Marks:")
print(df.mean(numeric_only=True))

# Add Total and Percentage
df["Total"] = df[["Math", "Science", "English"]].sum(axis=1)
df["Percentage"] = df["Total"] / 3

df
```

```python
import matplotlib.pyplot as plt

df.set_index("Name")[["Math", "Science", "English"]].plot(kind='bar')
plt.title("Student Marks Comparison")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
df["Percentage"].plot(kind='hist')
plt.title("Percentage Distribution")
plt.xlabel("Percentage")
plt.show()
```

Insights:
Highest scorer is Priya
Most students scored between 70–90%
Science has slightly higher average

Conclusion:
Data was cleaned and analyzed
Graphs helped understand student performance
